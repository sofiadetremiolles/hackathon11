```
# prompt: write me a code to find how many times each customer purchased the same product (ProductID), use index as transaction ID

# Calculate the purchase frequency of each product by each customer
customer_product_purchases = merged_df.groupby(['ClientID', 'ProductID']).size().reset_index(name='PurchaseCount')

# Display the result
customer_product_purchases
```

```
import matplotlib.pyplot as plt
import numpy as np

# Filter PurchaseCount values below 10
filtered_purchase_counts = customer_product_purchases[customer_product_purchases['PurchaseCount'] < 10]

# Create the histogram
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(filtered_purchase_counts['PurchaseCount'], bins=range(10), align='left', rwidth=0.8)

# Move the 0 on the x-axis to the y-axis
plt.xlim(left=0)

# Add percentages on top of each bar
total_counts = np.sum(counts)
for i, count in enumerate(counts):
    percentage = (count / total_counts) * 100
    plt.text(bins[i] + 0.4, count + 5, f'{percentage:.1f}%', ha='right', va='bottom', fontsize=12)  # Increased fontsize here

# Increase font size for labels and title
plt.xlabel('Purchase Count', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Purchase Counts (Below 10)', fontsize=16)
plt.xticks([0,1,2,3,4,5,6,7], fontsize=12)  # Ensure integer ticks on the x-axis and increase fontsize
plt.yticks(fontsize=12)  # Increase fontsize for y-axis ticks

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'customer_product_purchases' DataFrame is already created

purchase_counts = customer_product_purchases['PurchaseCount'].value_counts().sort_index()

# Combine purchase counts of 4 and above into a single category '4+'
purchase_counts_combined = purchase_counts.copy()
purchase_counts_combined['4+'] = purchase_counts[purchase_counts.index >= 4].sum()

# Convert index to integers (excluding '4+')
numerical_index = purchase_counts_combined.index[:-1].astype(int)

# Filter the original purchase_counts and update purchase_counts_combined
purchase_counts_combined = purchase_counts[purchase_counts.index.astype(int) < 4]

# Append the '4+' category back using concat
purchase_counts_combined = pd.concat([purchase_counts_combined, pd.Series({'4+': purchase_counts[purchase_counts.index.astype(int) >= 4].sum()}, index=['4+'])])


# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(purchase_counts_combined, labels=purchase_counts_combined.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Purchase Counts')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```
# prompt: write me a table of the total expenses per category per month for each year

import pandas as pd
# Convert 'TransactionDate' to datetime objects
fra_df['SaleTransactionDate'] = pd.to_datetime(fra_df['SaleTransactionDate'])

# Extract year and month
fra_df['Year'] = fra_df['SaleTransactionDate'].dt.year
fra_df['Month'] = fra_df['SaleTransactionDate'].dt.month

# Group by year, month, and category, then sum expenses
expenses_by_category = fra_df.groupby(['Year', 'Month', 'Category'])['SalesNetAmountEuro'].sum().reset_index()

# Display the table
expenses_by_category
```

```
# prompt: make a graph of lines with different color for every Category and with the x axis being time

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'expenses_by_category' DataFrame is already created as in your previous code

# Create the line plot
plt.figure(figsize=(12, 6))

for category in expenses_by_category['Category'].unique():
    category_data = expenses_by_category[expenses_by_category['Category'] == category]
    plt.plot(pd.to_datetime(category_data['Year'].astype(str) + '-' + category_data['Month'].astype(str), format='%Y-%m'), category_data['SalesNetAmountEuro'], label=category)

plt.xlabel('Time')
plt.ylabel('SalesNetAmountEuro')
plt.title('SalesNetAmountEuro by Category Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# prompt: make a graph of lines with different color for every Category and with the x axis being time but only with hockey, skating and skiing

import pandas as pd
import matplotlib.pyplot as plt
# Assuming 'expenses_by_category' DataFrame is already created as in your previous code

# Filter for 'Hockey', 'Skating', and 'Skiing' categories
filtered_expenses = expenses_by_category[expenses_by_category['Category'].isin(['Hockey', 'Skating', 'Skiing'])]

# Create the line plot
plt.figure(figsize=(12, 6))

for category in filtered_expenses['Category'].unique():
    category_data = filtered_expenses[filtered_expenses['Category'] == category]
    plt.plot(pd.to_datetime(category_data['Year'].astype(str) + '-' + category_data['Month'].astype(str), format='%Y-%m'), category_data['SalesNetAmountEuro'], label=category)

plt.xlabel('Time')
plt.ylabel('SalesNetAmountEuro')
plt.title('SalesNetAmountEuro for Hockey, Skating, and Skiing Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# prompt: make a table of the total sales per month per year grouped as below:
# Team Ball Sports
# 	Basketball
# 	Football
# 	Rugby
# Court & Net Ball Sports
# 	Volleyball
# 	Handball
# 	Tennis
# 	Badminton
# Stick & Puck Sports
# 	Golf
# 	Cricket
# 	Softball
# 	Baseball
# Winter Sports
# 	Skiing
# 	Skating
# 	Hockey
# Cycling & Other
# 	Cycling
# 	Running
# 	Beach Sports

# Assuming 'expenses_by_category' DataFrame is already created as in your previous code

# Create a dictionary to map categories to sports groups
sports_mapping = {
    'Basketball': 'Team Ball Sports',
    'Football': 'Team Ball Sports',
    'Rugby': 'Team Ball Sports',
    'Volleyball': 'Court & Net Ball Sports',
    'Handball': 'Court & Net Ball Sports',
    'Tennis': 'Court & Net Ball Sports',
    'Badminton': 'Court & Net Ball Sports',
    'Golf': 'Stick & Puck Sports',
    'Cricket': 'Stick & Puck Sports',
    'Softball': 'Stick & Puck Sports',
    'Baseball': 'Stick & Puck Sports',
    'Skiing': 'Winter Sports',
    'Skating': 'Winter Sports',
    'Hockey': 'Winter Sports',
    'Cycling': 'Cycling & Other',
    'Running': 'Cycling & Other',
    # Add other categories and their mappings here
}

# Add the sports group column
expenses_by_category['SportsGroup'] = expenses_by_category['Category'].map(sports_mapping)

# Group by year, month, and sports group, then sum sales
sales_by_sports_group = expenses_by_category.groupby(['Year', 'Month', 'SportsGroup'])['SalesNetAmountEuro'].sum().reset_index()

# Display the table
sales_by_sports_group
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'sales_by_sports_group' DataFrame is already created

# Create the line plot
plt.figure(figsize=(12, 6))

for sports_group in sales_by_sports_group['SportsGroup'].unique():
    sports_group_data = sales_by_sports_group[sales_by_sports_group['SportsGroup'] == sports_group]
    plt.plot(pd.to_datetime(sports_group_data['Year'].astype(str) + '-' + sports_group_data['Month'].astype(str), format='%Y-%m'), sports_group_data['SalesNetAmountEuro'], label=sports_group)

# Increase font size for all elements
plt.xlabel('Time', fontsize=14)
plt.ylabel('SalesNetAmountEuro', fontsize=14)
plt.title('SalesNetAmountEuro by Sports Group Over Time', fontsize=16)
plt.legend(fontsize=12)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Remove vertical gridlines and keep horizontal gridlines
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
